#### Principal Componenet Analysis

- PCA is a process of figuring out most important features or principal components that has the most impact on the target variable.
- A techique used in ML to reduce dimension.

**Exmples**
- lets say you are building a ML model to predict house 
prices from data with features(town, area, bathroom, plot, trees nearby, price). The featurs like "bathroom" and "trees nearby" effect the price very less or nothing.
- In problem of detecting numbers from 8x8 pixel images, there are 64 pixels in an image and some edge and corner pixels are always black and are not important feature.
- In real life model you can have hundereds of features in data. Thus, it is necessary to detect the important ones.

**What if we get rid of non important features?**
- Faster training and inference.
- Data visualtzation becomes easier.

**How it works?**
- we have to elminate the feature with very low variance. Like pixel_0, pixel_1, pixel_64 mostly have value = 0.0 for almost every number. Thus the variace of this feature is very low.
- PCA(n_components=10) will give the 10 features with most  in value, that are the central pixels.

**Few this to keep in mind before using PCA:**
- Scale features before applying PCA
- accuracy might drop


```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
dataset.keys()
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df
```

```python
x = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95) # retain 95% of usefull features

x_pca = pca.fit_transform(x)
x_pca.shape # 29 features left
```

```python
X_train_pca, X_test_pca, Y_train, Y_test = train_test_split(x_pca, y, test_size=0.2, random_state=30)
```

```python
model = LogisticRegression()
model.fit(X_train_pca, Y_train)
model.score(X_test_pca, Y_test)
```

the accuracy is comparmised a little but the computation required was very less.
