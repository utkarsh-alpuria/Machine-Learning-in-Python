#### Linear Regression

import necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

creating a dataframe from our excel file

```python
df = pd.read_excel("ML_Algo_Datasets/houseprice.xlsx")  # reading training dataset
df
```

Creating a scatter plot oh house price(US$) versus house area(sqft)

```python
%matplotlib inline
plt.xlabel("area(sqft)")
plt.ylabel("price(US$)")
plt.scatter(df.area, df.price, color="red", marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price) # trainging the model from dataset using linear regression algorithim
```

```python
reg.predict([[3300]]) # prediction of price according from area
```

y = mx + c  --> linear regression formulae

m --> coeficient

c --> intercept

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("area(sqft)")
plt.ylabel("price(US$)")
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[["area"]]), color="blue")
```

#### Linear Regression with Multiple Variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
homePrice_df = pd.read_excel("ML_Algo_Datasets/homePrice.xlsx")
homePrice_df
```

we need to re process this data since some values are missing

```python
import math
median_bedroom = math.floor(homePrice_df.bedroom.median())
median_bedroom
```

```python
homePrice_df.bedroom = homePrice_df.bedroom.fillna(median_bedroom)
homePrice_df
```

```python
reg2 = linear_model.LinearRegression()
reg2.fit(homePrice_df[["area", "bedroom", "age"]], homePrice_df.price)  # training model
```

Y = M1 * area + M2 * bedroom + M3 * age + C

```python
reg2.coef_
```

```python
reg2.intercept_
```

```python
reg2.predict([[3842, 2, 4]])
```
