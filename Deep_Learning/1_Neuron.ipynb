### What is a Neuron?
- lets take example of logistic regression
- Neuron has to functions
    1) create equation of weighted sum z = mx+b (M1X1 + M2X2 + M3X3...+ b)
    2) apply activation function (Sigmoid in this case) function

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
%matplotlib inline

insurance_df = pd.read_excel("..\ML_Algo_Datasets\Insurance.xlsx")
insurance_df.head(5)
```

```python
plt.scatter(insurance_df.age, insurance_df.bought_insurance, marker="+", color="red")
```

```python
# 1) lets do first step of creation equation z = mx + b 
reg = linear_model.LinearRegression()
reg.fit(insurance_df[["age"]], insurance_df.bought_insurance)
```

```python
%matplotlib inline
plt.xlabel("age")
plt.ylabel("bought_insurance")
plt.scatter(insurance_df.age, insurance_df.bought_insurance, color="red", marker="+")
plt.plot(insurance_df.age, reg.predict(insurance_df[["age"]]), color="blue")
```

```python
# value of m(weights) 
reg.coef_
```

```python
# value of b(bias)
reg.intercept_
```

```python
# 2) apply sigmoid function
import math
def sigmoid(x):
    return 1/(1 + math.exp(-x))

def prediction_function(age):
    z = 0.023 * age - 0.42 #(z = mx + b)
    y = sigmoid(z)
    return y #(y is between 0 t0 1)
```

```python
age = 20
prediction_function(age)
```

- Here we had only single neuron with single input(age)
- in NN we can have multiple inputs.
    - $y = w_1x_1 + w_2x_2 + w_3x_3 + b$
    - $y = \Sigma_{i=0}^n w_ix_i + b$
